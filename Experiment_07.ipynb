```python
P = np.polynomial.polynomial.Polynomial([-6.16,23.26,-16.84,-13.58,10.32,-19.98])
```

```python
R = P.roots()
```

```python
fnmlist = rf.fractional_newton_method(P,.01+ 0.j,1.01)
print(len(fnmlist))
```

```python
plt.figure(figsize=(8,8))
plt.scatter(fnmlist.real,fnmlist.imag,c=np.linspace(0,1,len(fnmlist)),alpha=.5)
plt.scatter(R.real,R.imag,alpha=.3,s=500)
plt.show()
```
