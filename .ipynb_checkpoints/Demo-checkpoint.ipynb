# Roots of a random polynomial

```python
params = [('normal',[0,1],{'size':101})]
coeffs = exp.random_coeffs(params)
```

```python
exp.poly_roots(coeffs)
```

# Roots of the its derivative

```python
exp.poly_deriv_roots(coeffs,dx=2)
```

```python
exp.poly_deriv_roots(coeffs,dx=15)
```

```python
exp.poly_deriv_roots(coeffs,dx=40)
```

```python
exp.poly_deriv_roots(coeffs,dx=70)
```

This illustrates the Gauss-Lucas Theorem: the roots of the derivative of a polynomial are in the convex hull of the roots of the original polynomial.
