# Root Finding with Fractional Calculus

**Problem**

Usually finding roots of a polynomial is done by forming the companion matrix and solving for the eigenvalues. This is bad though when $n$ is very large or if the polynomial is multi-dimensional. There is a relationship between the roots of a polynomial $p(x)$ and the roots of its derivatives $p'(x),p''(x),\ldots,p^{(n)}(x)$. It might be possible to find roots by differentiating the polynomial $p(x)$ down to some a derivative polynomial where the roots are known, and then keep track of the roots while antidifferentiating back up to the original polynomial.

# Wilkinson Polynomial

because curiosity

```python
roots = np.arange(1,21)

# the coefficients defining wilkinson's polynomial
coeffs = np.polynomial.polynomial.polyfromroots(roots)
```

```python
rf.timelapse(coeffs,0,2,20)
```

```python
rf.animate_roots(coeffs,0,18,n_frames=250,t_interval=100,plot_range=[(-.5,20),(-5,5)])
```

The following plot shows the continuous deformation of the roots from the 14th to the 18th derivative. The slower speed and larger number of frames shows exactly how the roots 'collide' and disappear. This explains the behavior of the above animation.

```python
rf.animate_roots(coeffs,14,18,n_frames=500,t_interval=150,plot_range=[(-.5,20),(-5,5)])
```

```python
coeffs = np.polynomial.polynomial.polyfromroots(np.ones(20))
rf.animate_roots(coeffs,0,18,n_frames=250,t_interval=100,plot_range=[(-0.1,2.2),(-1.1,1.1)])
```
