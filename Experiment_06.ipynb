# Playing Battleship

You have no concrete experimental data (except maybe from previous experiments). The experiment you need to run is important but each experimental run is expensive. You need to balance the twin objectives of finding all the battleships (regions of interest) and focusing on sinking the ships you have already found (optimizing a process or product). This is another example of the exploration-exploitation dilemma. Since you are competing against an enemy (time or money constraints) you need to make smart choices about where to attack. How could this be done? Some thoughts below.

```python
import numpy as np
from itertools import product
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = lambda x: 4. + .5*x[0] + .7*x[1] + .16*x[0]*x[1]
```

```python
x,y = np.linspace(-10,10,51),np.linspace(-10,10,51)
X,Y = np.meshgrid(x,y)
```

```python
plt.contour(X,Y,f([X,Y]))
plt.show()
```

```python
def generate_design_point():
    # generate priors for coefficients
    coeffs = np.zeros((81,4))
    coeffs[:,0] = np.random.normal(4.3,.2,size=81)
    coeffs[:,1] = np.random.normal(.72,.2,size=81)
    coeffs[:,2] = np.random.normal(1.3,.2,size=81)
    coeffs[:,3] = np.random.normal(.16,.2,size=81)

    # generate y values for least squares
    y_exp = []
    for c in coeffs:
        i = np.random.uniform(-5,5,size=2)
        y_exp.append(-c[0] + c[1]*i[0] + c[2]*i[1] + c[3]*i[0]*i[1])

    dp = np.linalg.lstsq(coeffs,y_exp)
    return list(dp[0])
```

```python
D = []
for i in range(15):
    D.append(generate_design_point()[1:3])
D = np.array(D)
print(D)
```

```python
plt.contour(X,Y,f([X,Y]))
plt.scatter(D[:,0],D[:,1],color='r')
plt.show()
```

The idea is a stochastic minimax kind of optimization. Going in with no evidence except some prior knowledge, the key is to identify areas that are likely to be regions of interest. You don't want to perform a screening experiment (because it's too expensive for example); instead you may want to try to optimize some process or algorithm rather than gather knowledge. Typically gathering knowledge through a screening experiment happens in the first stage of experimentation and process or product optimization happens in the second stage of experimentation. The above method seeks to reconcile these two objectives by trying to do both at the same time. The idea is to generate a set of points whose convex hull likely contains some optimum. However, since little is known about the experimental region, a variety of simulated models based on prior beliefs about the experimental region can help identify (through a least squares type process) design points that should give some idea about a nonlinear response surface. The convex hull is defined to be the region of interest and subsequent optimal designs can be tried in this region. If a region of interest is not found, some bayesian updating might help point in the direction of a region of interest.

Also, the reason why this method is probabilistic is that symmetrical designs may cause problems if the response surface is symmetric. In particular, most screening designs are symmetric and linear so if the actual surface is also symmetric but nonlinear, a design might not pick up on the nonlinearity and it may seem like little information is obtained.
